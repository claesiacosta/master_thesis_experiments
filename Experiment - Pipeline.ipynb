```python
class DerivationalAnalyzer:
    def __init__(self):
        self.suffix_patterns = {
            "verb to noun": "ção, mento, tório, dor",
            "adj to noun": "ice, dade",
            "verb to adj": "vel, nte",
            "noun to adj": "ense, ar",
            "noun, adj to verb": "izar, escer",
            "adj to adv": "mente"
        }
    
    def identify_derivational_class(self, word, pos):
        for deriv_class, suffixs in self.suffix_patterns.items():
            for suffix in suffixs.split(", "):
                if word.endswith(suffix) and pos.lower() == deriv_class.split(" ")[-1]:
                    return deriv_class, word[:-len(suffix)]
        
        return None, None
```

```python
analyzer = DerivationalAnalyzer()
# Aplicar a análise derivacional e criar colunas para os resultados

portilexicon_filtered[['derivational_class', 'base_word']] = portilexicon_filtered.apply(lambda row: pd.Series(analyzer.identify_derivational_class(row['words'], row['pos'])), axis=1)
```

```python
# Itere pelas linhas do dataset_origem
for index, row in new_merged_dataset.iterrows():
    # Obtém a palavra de referência da linha atual do dataset_origem
    palavra_referencia = row[2]  # Substitua 'coluna_referencia' pelo nome da coluna apropriada

    # Encontre o índice da linha no dataset_destino que contém a palavra de referência
    indice_referencia = nomlexpt[nomlexpt[2] == palavra_referencia].index[0]

    # Divida o dataset_destino em duas partes: antes e depois da linha de referência
    parte_superior = nomlexpt.iloc[:indice_referencia + 1]
    parte_inferior = nomlexpt.iloc[indice_referencia + 1:]

    # Crie um DataFrame temporário com a linha atual do dataset_origem
    nova_linha = pd.DataFrame([[None, row['word_pos'], row['words'], row['pos'], None, None, None, 'Type=Derivation', None, None, row['TextoStemmed']]])

    # Concatene as partes superior, nova_linha e parte inferior para criar o novo dataset_destino
    nomlexpt = pd.concat([parte_superior, nova_linha, parte_inferior], ignore_index=True)

# Agora, dataset_destino conterá as linhas adicionadas nas posições corretas com base nas palavras de referência.
```
